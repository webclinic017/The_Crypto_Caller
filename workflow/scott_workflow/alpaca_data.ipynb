```python
import config, requests, json
from pathlib import Path
import pandas as pd
from datetime import datetime, date
import alpaca_trade_api as tradeapi
# config is where the ALPACA API KEYS are stored
```

```python
# create a webscraper to dump a list of cryptos from coinbase into a csv file
holdings = pd.read_csv(
    Path("datasets/crypto_df.csv")
)

holdings.head()
```

```python
holdings = holdings['symbol']
holdings.head()
```

```python
symbols = []

for symbol in holdings:
    symbols.append(symbol)

print(symbols)
```

```python
today = date.today()

def sub_years(years):
    today = date.today()
    try:
        return today.replace(year = today.year - years)
    except ValueError:
        return today + (date(today.year + years, 1, 1) - date(today.year, 1, 1))

start = sub_years(1).strftime('%Y-%m-%dT%H:%M:%SZ')

end = datetime.utcnow()
end = end.isoformat("T") + "Z"

print(start)
print(end)
```

```python
alpaca = tradeapi.REST(
    config.API_KEY,
    config.SECRET_KEY,
    api_version="v2"
)
```

```python
stocks = ['BTCUSD', 'ETHUSD', 'BCHUSD', 'LTCUSD']
```

```python
crpyto_df = alpaca.get_barset(
    stocks,
    timeframe='1D', 
    start=start,
    end=end
).df

crpyto_df.tail()
```

```python
BTC = 'BTCUSD'
ETH = 'ETHUSD'
#minute_bar_url = config.BARS_URL + '/5min?symbols=MSFT&limit=1000'

day_bars_url = '{}/{}/bars?exchanges=ERSX&start={}&end={}&limit=1000&timeframe=1Day'.format(config.BARS_URL, stocks, start, end)

r = requests.get(day_bars_url, headers=config.HEADERS)
```

```python
list = []
for info in r.json()['bars']:
    list.append(pd.Series(info))

test_df = pd.DataFrame(data=list)
test_df.head()
```

```python
test_df.rename(columns={'t':'Date', 'x':'exchange', 'o':'open', 'h':'high', 'l':'low', 'c':'close', 'v':'volume', 'n':'number_of_trades', 'vw':'volume_weighted_average_price'})
```

```python
stocks = ['BTCUSD', 'ETHUSD', 'BCHUSD', 'LTCUSD']

for stock in stocks:

    day_bars_url = '{}/{}/bars?exchanges=ERSX&start={}&end={}&limit=10000&timeframe=1Min'.format(config.BARS_URL, stock, start, end)

    r = requests.get(day_bars_url, headers=config.HEADERS)

    list = []
    for info in r.json()['bars']:
        list.append(pd.Series(info))

    test_df = pd.DataFrame(data=list)
    test_df = test_df.rename(columns={'t':'date', 'x':'exchange', 'o':'open', 'h':'high', 'l':'low', 'c':'close', 'v':'volume', 'n':'number_of_trades', 'vw':'volume_weighted_average_price'})
    test_df.to_csv('datasets/ohlc/{}.csv'.format(stock))
```
