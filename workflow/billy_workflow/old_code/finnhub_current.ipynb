```python
import finnhub
import pandas as pd
import os
from dotenv import load_dotenv
import json
import datetime
from datetime import datetime, timedelta
import time

load_dotenv()
finnhub_api_key = os.getenv("FINNHUB_API_KEY")
finnhub_client = finnhub.Client(api_key=finnhub_api_key)
```

```python
crypto_list = finnhub_client.crypto_symbols('COINBASE')
crypto_df = pd.DataFrame(crypto_list, columns=["symbol"])
crypto_df
```

```python
crypto_df = crypto_df.str.replace("COINBASE:", "")
crypto_df
```

```python
symbol_list = []
for row in crypto_df:
    symbol_list.append(pd.Series(row))
symbol_list
```
