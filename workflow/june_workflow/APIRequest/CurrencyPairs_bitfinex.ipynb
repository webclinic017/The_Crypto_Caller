```python
import bitfinex
import pandas as pd
from pathlib import Path

# Create api instance
api_v1 = bitfinex.bitfinex_v1.api_v1()
pairs = api_v1.symbols()
pairs
```

```python
df = []

for crypto in pairs:
    if crypto[-1]=="d":
        df.append(crypto)

df = pd.DataFrame(data=df)
df.tail()
```

```python
# Create dataframe and save as csv file
names = ['symbol']
df = pd.DataFrame(pairs, columns=names)
df.to_csv(Path('../Resources/CurrencyPair.csv'))
```
